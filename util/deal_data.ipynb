```python
import json

def csv_to_json(csv_file: str, json_file: str) -> None:
    """
    将CSV文件转换为JSON格式，并保存到指定路径。
    
    Args:
        csv_file (str): 输入的CSV文件路径
        json_file (str): 输出的JSON文件路径
    
    示例:
        csv_to_json('data.csv', 'output.json')
    """
    # 读取CSV文件为DataFrame
    df = pd.read_csv(csv_file)
    
    # 将DataFrame转换为JSON格式并写入文件
    df.to_json(json_file, orient='records', force_ascii=False, indent=4)

    print(f"CSV文件成功转换为JSON并保存到: {json_file}")
    

def csv_to_jsonl(csv_file: str, jsonl_file: str) -> None:
    """
    将CSV文件转换为JSONL格式，并保存到指定路径。
    
    Args:
        csv_file (str): 输入的CSV文件路径
        jsonl_file (str): 输出的JSONL文件路径
    
    示例:
        csv_to_jsonl('data.csv', 'output.jsonl')
    """
    # 读取CSV文件为DataFrame
    df = pd.read_csv(csv_file)
    
    # 打开JSONL文件进行写入
    with open(jsonl_file, 'w', encoding='utf-8') as f:
        # 逐行将DataFrame转换为JSON格式并写入文件
        for record in df.to_dict(orient='records'):
            json_record = json.dumps(record, ensure_ascii=False)  # 转换为合法的JSON格式
            f.write(f"{json_record}\n")

    print(f"CSV文件成功转换为JSONL并保存到: {jsonl_file}")
```

```python
import pandas as pd
import cpca
from pyecharts.charts import Map
from pyecharts import options as opts
from pyecharts.globals import GeoType
from pyecharts.render import make_snapshot
from snapshot_selenium import snapshot

def extract_location_counts(data: pd.DataFrame, fields=['title', 'content']) -> dict:
    """
    使用cpca提取省、市、县区信息，并统计出现次数。
    Args:
        data: 包含文本内容的数据框。
        fields: 要提取的文本字段列表。
    Returns:
        包含省、市、县区统计结果的字典。
    """
    texts = data[fields].apply(lambda x: '\n'.join(x), axis=1)
    location_data = cpca.transform(texts.tolist())

    # 统计省份数量
    province_counts = location_data['省'].value_counts().reset_index()
    province_counts.columns = ['Province', 'Count']

    # 统计城市数量
    city_counts = location_data['市'].value_counts().reset_index()
    city_counts.columns = ['City', 'Count']

    # 统计县区数量
    county_counts = location_data['区'].value_counts().reset_index()
    county_counts.columns = ['County', 'Count']

    return {
        'province_counts': province_counts,
        'city_counts': city_counts,
        'county_counts': county_counts
    }

def echarts_cate_draw(counts_df: pd.DataFrame, area_column: str, value_column: str, png_path: str):
    """
    使用pyecharts生成热力图，并保存为PNG文件。
    Args:
        counts_df: 包含地区和统计值的数据框。
        area_column: 地区列的列名（例如 'Province'、'City'、'County'）。
        value_column: 统计值的列名（例如 'Count'）。
        png_path: 保存PNG文件的路径。
    """
    # 创建地图对象
    c = (
        Map()
        .add(
            series_name="",
            data_pair=[(row[area_column], row[value_column]) for idx, row in counts_df.iterrows()],
            maptype="china",
            is_roam=True,
            label_opts=opts.LabelOpts(font_size=8)  # 调整标签字体大小
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="地域热力图"),
            visualmap_opts=opts.VisualMapOpts(
                max_=counts_df[value_column].max(),
                is_piecewise=True,
                textstyle_opts=opts.TextStyleOpts(font_size=10)
            )
        )
    )

    # 渲染并保存为PNG
    c.render("temp_render.html")
    make_snapshot(snapshot, "temp_render.html", png_path)

# 示例数据
data = pd.DataFrame({
    'title': ['标题1', '标题2', '标题3'],
    'content': ['北京市海淀区中关村', '上海市浦东新区陆家嘴', '广东省深圳市南山区']
})

# 提取地点统计信息
location_counts = extract_location_counts(data)

# 绘制省份热力图并保存为 PNG
echarts_cate_draw(
    counts_df=location_counts['province_counts'],
    area_column='Province',
    value_column='Count',
    png_path='province_heatmap.png'
)
```

```python
from pyecharts.charts import Map, Geo
from pyecharts import options as opts
from pyecharts.globals import ChartType
from typing import Dict, List, Tuple

def normalize_size(value: int, min_value: int, max_value: int, min_size: int, max_size: int) -> int:
    """将计数值归一化为合适的圆圈大小
    
    Args:
        value: 当前城市的计数值
        min_value: 所有城市计数中的最小值
        max_value: 所有城市计数中的最大值
        min_size: 最小圆圈大小
        max_size: 最大圆圈大小
    
    Returns:
        int: 映射后的圆圈大小
    """
    if max_value == min_value:
        return (min_size + max_size) // 2  # 若所有值相等，则使用中等大小
    return int(min_size + (value - min_value) * (max_size - min_size) / (max_value - min_value))

def generate_heatmap(province_data: Dict[str, int], city_data: Dict[str, int]) -> None:
    """生成中国省份和城市的热力图

    Args:
        province_data: 映射省份到计数的字典
        city_data: 映射城市到计数的字典
    """
    
    # 创建中国地图
    map_chart = Map(init_opts=opts.InitOpts(width="1000px", height="800px"))
    
    # 将省份的数据加入地图
    provinces = list(province_data.keys())
    province_counts = list(province_data.values())
    
    map_chart.add(
        "省份热力图", 
        [list(z) for z in zip(provinces, province_counts)], 
        maptype="china",
        is_map_symbol_show=False  # 关闭默认的小红点
    )
    
    # 定义颜色和视觉映射
    map_chart.set_global_opts(
        title_opts=opts.TitleOpts(title="中国省份与城市热力图"),
        visualmap_opts=opts.VisualMapOpts(
            max_=max(province_counts),  # 最大值
            min_=min(province_counts),  # 最小值
            is_piecewise=True,  # 分段显示
            pieces=[
                {"min": 1000, "label": ">=1000", "color": "#FF0000"},
                {"min": 500, "max": 999, "label": "500-999", "color": "#FF4500"},
                # {"min": 100, "max": 499", "label": "100-499", "color": "#FFA500"},
                {"min": 1, "max": 99, "label": "1-99", "color": "#FFFF00"},
            ]
        )
    )
    
    # 创建城市散点图，使用 Geo 来显示城市数据
    geo_chart = Geo(init_opts=opts.InitOpts(width="1000px", height="800px"))
    
    # 获取城市数据
    cities = list(city_data.keys())
    city_counts = list(city_data.values())
    
    # 获取最大和最小计数值
    max_count = max(city_counts)
    min_count = min(city_counts)
    
    # 根据城市的计数值计算圆圈大小
    scatter_data = [
        (city, count, normalize_size(count, min_count, max_count, min_size=5, max_size=20))
        for city, count in zip(cities, city_counts)
    ]
    
    # 将城市的数据加入 Geo 图层
    geo_chart.add_schema(maptype="china")
    geo_chart.add(
        "城市热力图",
        [(data[0], data[1]) for data in scatter_data],  # 城市名和计数值
        type_=ChartType.EFFECT_SCATTER,  # 使用散点显示城市数据
        symbol_size=[data[2] for data in scatter_data]  # 根据计数值设置散点大小
    )
    
    # 定义城市数据的视觉效果
    geo_chart.set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),  # 不显示地理标签
    )
    geo_chart.set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            max_=max(city_counts),
            min_=min(city_counts),
            is_piecewise=True
        )
    )
    
    # 渲染省份和城市两个图表
    map_chart.render("province_heatmap.html")
    geo_chart.render("city_heatmap.html")
    print("热力图已生成，请在当前文件夹中查看 `province_heatmap.html` 和 `city_heatmap.html` 文件。")

# 示例数据
province_data_example = {
    "北京": 1200,
    "上海": 800,
    "广东": 500,
    "浙江": 300,
    "四川": 100
}

city_data_example = {
    "北京": 1200,
    "上海": 800,
    "广州": 400,
    "杭州": 250,
    "成都": 150
}

# 调用函数生成热力图
generate_heatmap(province_data_example, city_data_example)
```

```python
import folium
from folium.plugins import HeatMap
import pandas as pd

# 示例数据：城市的经纬度和计数
data = [
    {"city": "北京", "lat": 39.9042, "lon": 116.4074, "count": 1200},
    {"city": "上海", "lat": 31.2304, "lon": 121.4737, "count": 800},
    {"city": "广州", "lat": 23.1291, "lon": 113.2644, "count": 400},
    {"city": "杭州", "lat": 30.2741, "lon": 120.1551, "count": 250},
    {"city": "成都", "lat": 30.5728, "lon": 104.0668, "count": 150}
]

# 转换为 DataFrame 方便处理
df = pd.DataFrame(data)

# 创建 Folium 地图对象
m = folium.Map(location=[35.8617, 104.1954], zoom_start=5)  # 以中国为中心

# 构建热力图数据：需要提供[纬度，经度，权重]的列表
heat_data = [[row['lat'], row['lon'], row['count']] for index, row in df.iterrows()]

# 将热力图数据添加到地图上
HeatMap(heat_data).add_to(m)

# 保存生成的地图为 HTML 文件
m.save("processed.html")
print("热力图已生成，文件名为 processed.html")
```

```python
import folium
from folium.plugins import HeatMap
import cpca
import pandas as pd
from typing import Dict

# 从 cpca 代码中导入 ad_2_addr_dict
from cpca import _init_data

ad_2_addr_dict, matcher = _init_data()  # 初始化 adcode 数据和 matcher

def get_lat_lon(city: str) -> tuple:
    """
    使用 cpca 获取城市的经纬度信息，如果无法匹配则返回 (None, None)
    """
    # 使用 cpca 提取省市区信息
    location_df = cpca.transform([city], pos_sensitive=True)
    
    # 检查 DataFrame 是否为空
    if location_df.empty:
        print(f"城市 {city} 无法解析，跳过。")
        return None, None
    
    # 尝试获取 adcode
    adcode = location_df.iloc[0].get('adcode')
    if pd.isna(adcode):
        print(f"城市 {city} 无法获取 adcode，跳过。")
        return None, None
    
    # 从 adcode 获取地址信息
    addr_info = ad_2_addr_dict.get(adcode.ljust(12, '0'))
    if addr_info:
        return float(addr_info.latitude), float(addr_info.longitude)
    else:
        print(f"城市 {city} 的 adcode 未找到匹配的地址信息，跳过。")
        return None, None

def create_heatmap(city_count: Dict[str, int]) -> str:
    """
    创建热力图并返回 HTML 文件路径
    
    Args:
        city_count (Dict[str, int]): 映射城市名称到计数值的字典
    
    Returns:
        str: 生成的 HTML 文件路径
    """
    data = []
    
    # 遍历每个城市及其计数
    for city, count in city_count.items():
        # 获取城市的经纬度信息
        lat, lon = get_lat_lon(city)
        
        if lat is None or lon is None:
            print(f"城市 {city} 无法找到经纬度，跳过。")
            continue
        
        # 将城市的经纬度及计数信息加入列表
        data.append([lat, lon, count])
    
    # 创建 Folium 地图对象
    m = folium.Map(location=[35.8617, 104.1954], zoom_start=5)  # 以中国为中心
    
    # 将热力图数据添加到地图上
    HeatMap(data).add_to(m)
    
    # 保存生成的地图为 HTML 文件
    output_path = "city_heatmap.html"
    m.save(output_path)
    
    return output_path

# 示例数据
city_count_example = {
    "北京市": 1200,
    "上海市": 800,
    "江西省": 400,
    "湖南省": 250,
    "成都市": 150,
    "不存在的城市": 100  # 测试无法找到经纬度的情况
}

# 生成热力图
html_path = create_heatmap(city_count_example)
print(f"热力图已生成，文件路径为: {html_path}")
```
