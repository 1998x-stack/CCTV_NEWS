```python
import pandas as pd
df = pd.read_csv('video_data.csv')
df.head()
```

```python
df = df[~df['content'].str.contains('内容未找到')]
df['title'] = df['title'].str.strip("[视频]")
df['content'] = df['content'].str.strip("央视网消息（新闻联播）：")
df['title'] = df['title'].str.strip()
df['content'] = df['content'].str.strip()
df = df[['date', 'duration', 'title', 'content', 'link']]
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df = df.sort_values(by='date', ascending=True)
df = df.reset_index(drop=True)
df.head()
```

```python
df.to_csv('data.csv')
```

```python
df_kuai_xun = df[df['title'] == '国内联播快讯']
df_kuai_xun = df_kuai_xun.reset_index(drop=True)
df_kuai_xun.head()
```
